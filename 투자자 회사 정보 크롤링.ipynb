# 투자자 회사 정보 크롤링
* 한국 스타트업 투자 데이터베이스 사이트인 https://thevc.kr/를 활용합니다.
* 한국 기업들에 투자하는 전세계 Top10 투자자 회사들의 회사명, 국적, 분류, 총 투자 건수, 총 투자 집행 금액을 크롤링합니다.
* selenium을 활용한 웹크롤링을 통해 데이터를 수집합니다.

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

driver_path = "/Users/alexnam/Downloads/chromedriver"
driver = webdriver.Chrome(driver_path)

driver.get("https://thevc.kr/")

#로그인 버튼 클릭
time.sleep(3)
login_button="#layout-wrap > div.vc-element-header > div.always-showing > header > div > div.right > a"
driver.find_element_by_css_selector(login_button).click()
time.sleep(3)

#email 입력
elem = driver.find_element_by_id("input-email")
elem.clear()
elem.send_keys("alex8030@naver.com")

time.sleep(2)

#password 입력
elem = driver.find_element_by_id("input-password")
elem.clear()
elem.send_keys("nam56368133!")

# Enter 입력
time.sleep(2)      
elem.send_keys(Keys.RETURN)

# 투자자 버튼 클릭
time.sleep(3)
vc_list_button = '#layout-wrap > div.vc-element-header > div.always-showing > nav > div > div > a:nth-child(3)'
driver.find_element_by_css_selector(vc_list_button).click()

# 총 투자 집행 금액 내림차 순으로 정렬
time.sleep(3)
sort_button1 = '#layout-wrap > div.main-el > div > div:nth-child(2) > div.vc-browser > div.table-wrap > table > thead > tr > th:nth-child(5)'
#sort_button1 = '#layout-wrap > div.main-el > div > div:nth-child(2) > div.vc-browser > div.table-wrap > table > thead > tr > th:nth-child(4) > span:nth-child(2) > i'
driver.find_element_by_css_selector(sort_button1).click()
time.sleep(1)
sort_button2 = '#layout-wrap > div.main-el > div > div:nth-child(2) > div.vc-browser > div:nth-child(2) > div > section > header > button'
driver.find_element_by_css_selector(sort_button2).click()
time.sleep(1)
driver.find_element_by_css_selector(sort_button2).click()
time.sleep(1)

# 적용버튼 클릭
apply_button="#layout-wrap > div.main-el > div > div:nth-child(2) > div.vc-browser > div:nth-child(2) > div > section > footer > button:nth-child(3)"
driver.find_element_by_css_selector(apply_button).click()
time.sleep(2)
html=driver.page_source
print(html)
```

```python
# class로 분류
from bs4 import BeautifulSoup

soup = BeautifulSoup(html)
store = soup.select('.vc-typo-ellipsis')
store
```
