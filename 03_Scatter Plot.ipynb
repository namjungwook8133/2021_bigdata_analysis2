# 3.Scatter Plot
* 좌표상의 점들을 표시함으로써 두 변수간의 관계를 나타내줌

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(50) # [0,1) (0이상 1미만의 난수를 생성)
y = np.random.rand(50)
plt.scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x,y)
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
plt.show()
```

```python
# Color 지정
plt.scatter(x,y,c='r')
```

```python
# 점마다 다른 color 지정하기
x=[1,2,3]
y=[1,2,3]
colors=['r','g','b']
plt.scatter(x,y,c=colors)
plt.show()
```

```python
# ColorMap(색상지도)
plt.colormaps()
```

```python
x =np.random.rand(50)
y =np.random.rand(50)
plt.scatter(x,y, c=x,cmap='summer')
plt.colorbar()
plt.show()
```

```python
# Marker Size 변경하기
size = np.random.rand(50) * 100 #0.1*100 = 10
plt.scatter(x,y,c =x,cmap = 'summer',s = size) #'s = '는 데이터의 사이즈를 변경
plt.show()
```

```python
# 투명도 (alpha, 0~1의 값)
plt.scatter(x,y,c =x,cmap = 'summer',s = size,alpha=0.5)
plt.show()
```
