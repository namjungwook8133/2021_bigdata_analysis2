# 5. Pie Chart
* 백분율로 표시
* 데이터의 비율을 알아내는데 유용

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=[1,2,3,4]
plt.pie(data)
plt.show()
```

```python
labels =['치킨','핫바','족발','팅글']
plt.pie(data,labels = labels)
plt.show()
```

```python
plt.pie(data, labels=labels, autopct="%.1f%%")
plt.show()
```

```python
explode=[0.3,0,0,0]
plt.pie(data, labels=labels, autopct="%.1f%%", explode=explode, shadow=True)
plt.show()
```

```python
colors =['r', 'g', 'b', 'pink']
plt.pie(data, labels=labels, autopct="%.1f%%", explode=explode, shadow=True, colors=colors)
plt.show()
```
